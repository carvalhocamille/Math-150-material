```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x0 = np.random.rand(200) 
t = np.linspace(0.1, 20, 200, endpoint=False)
x = x0 + np.exp(2)*np.power(t,-1)
print(x0)
fig, ax = plt.subplots()
ax.plot(t,x)
```

```python
df = pd.DataFrame.from_dict({'t':t, 'x':x, 'x0':x0})
df.set_index('t')
df.to_csv('hw1_data.csv', index=False)
```

```python
file = 'hw1_data.csv'
df = pd.read_csv(file)
t = df.iloc[:,0]
x = df.iloc[:,1]
x0 = df.iloc[:,2]
fig, ax = plt.subplots()
ax.plot(t,x,'x')
```
