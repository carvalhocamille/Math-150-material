```python
#Recommanded libraries to import at the beginning of each Jupyter notebook
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Loading file data

```python
file = 'hw1_data.csv'
df = pd.read_csv(file)
t = df.iloc[:,0]
x = df.iloc[:,1]
x0 = df.iloc[:,2]
fig, ax = plt.subplots(1, 1, figsize=(10,10))
ax.plot(t,x,'x', label = 'x')
ax.plot(t, x0,'+', label = 'x0')
plt.title('Observations and predictions')
ax.set_xlabel("t")
ax.set_ylabel("x values")
ax.legend()
```

### Analyze error: plot absolute error

### Analyze error: plot log error

```python
fig, ax = plt.subplots(1, 1, figsize=(10,10))
ax.loglog(t,np.abs(x-x0), label = 'error')
ax.loglog(t,1/t,'r--', label = 'slope -1')
ax.legend()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))
ax1.semilogy(t,np.abs(x-x0), label = 'error semilogy')
ax1.legend()
ax2.semilogx(t,np.abs(x-x0), label = 'error semilogx')
ax2.legend()
```
