```python
from matplotlib.pyplot import xkcd
import numpy as np
import math
import matplotlib.pyplot as plt
```

# Logistic growth model

```python
f = lambda C: α*C*(1 - C/M)
α = 2
M = 10
t = np.arange(-2,12, 0.1)
with plt.xkcd():
    fig = plt.figure(figsize=(8, 8))
    ax = fig.add_subplot(1, 1, 1)
    ax.spines['bottom'].set_position(('data', 0.0))
    ax.spines['left'].set_position(('data', 0.0))
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    #ax.set_xlabel(loc = 'right')
    plt.plot(t, f(t),zorder = 0, label = 'logistic')
    plt.xlabel('t', loc ='right')
    plt.ylabel('f(C)', loc= 'top')
    plt.legend()
    plt.margins(0.1)
    
```

###### 

# Solution of the logistic growth for various parameters
### 1) for various rates

```python
C = lambda t, M, α: (2*M)/(M-2) * np.exp(α *t)/(1 + 2/(M-2)*np.exp(α *t))
α = 2
M = 10
t = np.arange(-2,10, 0.1)
with plt.xkcd():
    fig = plt.figure(figsize=(8, 8))
    ax = fig.add_subplot(1, 1, 1)
    ax.spines['bottom'].set_position(('data', 0.0))
    ax.spines['left'].set_position(('data', 0.0))
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    #ax.set_xlabel(loc = 'right')
    plt.plot(t, M + 0* C(t,M,α),'-.r')
    plt.plot(t, C(t,M,α-1),zorder = 0, label = f'α = {α-1}')
    plt.plot(t, C(t,M,α),zorder = 0, label = f'α = {α}')
    plt.plot(t, C(t,M,α+1),zorder = 0, label = f'α = {α+1}')
    plt.xlabel('t', loc ='right')
    #plt.ylabel('C(t)', loc= 'top')
    plt.legend(loc = 'center right')
    plt.margins(0.1)
```

### 2) For various carrying capacities

```python
with plt.xkcd():
    fig = plt.figure(figsize=(8, 8))
    ax = fig.add_subplot(1, 1, 1)
    ax.spines['bottom'].set_position(('data', 0.0))
    ax.spines['left'].set_position(('data', 0.0))
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    #ax.set_xlabel(loc = 'right')
    plt.plot(t, M + 0* C(t,M,α),'-.r')
    plt.plot(t, M -1+ 0* C(t,M,α),'-.r')
    plt.plot(t, M +1 + 0* C(t,M,α),'-.r')
    plt.plot(t, C(t,M-1,α),zorder = 0, label = f'M = {M-1}')
    plt.plot(t, C(t,M,α),zorder = 0, label = f'M = {M}')
    plt.plot(t, C(t,M+1,α),zorder = 0, label = f'M = {M+1}')
    plt.xlabel('t', loc ='right')
    #plt.ylabel('C(t)', loc= 'top')
    plt.legend(loc = 'center right')
    plt.margins(0.1)
```

### 3) For various initial conditions

```python
C = lambda t, M, α, c0: (c0*M)/(M-c0) * np.exp(α *t)/(1 + c0/(M-c0)*np.exp(α *t))
α = 2
M = 10
c0 = 2
t = np.arange(0,10, 0.1)
with plt.xkcd():
    fig = plt.figure(figsize=(8, 8))
    ax = fig.add_subplot(1, 1, 1)
    ax.spines['bottom'].set_position(('data', 0.0))
    ax.spines['left'].set_position(('data', 0.0))
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    #ax.set_xlabel(loc = 'right')
    plt.plot(t, M + 0* C(t,M,α, c0),'-.r')
    plt.plot(t, C(t,M,α, c0),zorder = 0, label = f'c0 = {c0}')
    plt.plot(t, C(t,M,α, M+2),zorder = 0, label = f'c0 = {M+2}')
    plt.plot(t, C(t,M,α, M-4),zorder = 0, label = f'c0 = {M-4}')
    plt.xlabel('t', loc ='right')
    #plt.ylabel('C(t)', loc= 'top')
    plt.legend(loc = 'center right')
    plt.margins(0.1)
```
