We consider here the discrete scheme
$$y_n = g(y_n)$$
with $$g(y_n) = \alpha y_n (1 - y_n)$$

```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
import sympy as sp
```

```python
from ipywidgets import interact 
@interact
def interact_sir(α:(0, 10, 0.2)=1, x0:(0, 1, 0.1)=0.2, N:(1, 100, 1)=10):
    g = lambda x: α*x *(1 - x**2) # g = @(x) alaph*x(1-x)
    x = np.linspace(-10, 10, 300)
    fig, ax = plt.subplots(1, 1, figsize=(10, 10))
    ax.plot(x, x, 'C0', label = 'y=x')
    ax.plot(x, g(x), 'C1', label = 'y = g(x)')
    ax.hlines(0, -10, 10, colors='k') 
    ax.vlines(0, -10, 10, colors='k') 
    for i in range(N):
        tmp = g(x0)
        ax.vlines(x0, x0, tmp, colors='C2') 
        ax.hlines(tmp, x0, tmp, colors='C2')
        x0 = tmp
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_ylim(-1.5,1.5)
    ax.set_xlim(-1.5,1.5)
    ax.grid(True)
    legend = ax.legend()
```
